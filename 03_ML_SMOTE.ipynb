# SMOTE(Synthetic Minority Oversampling Technique)

```python
# SMOTE is a technique used in  machine learning to address imbalance datasets where 
```

```python


from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000 ,n_redundant=0,n_features=2 ,n_clusters_per_class=1 , weights=[0.9] , random_state = 12 )
```

```python
len(X)
```

```python
len(y)
```

```python
import pandas as pd

df1 =  pd.DataFrame(X , columns = ["f1", "f2"])
df2 = pd.DataFrame(y , columns = ["target"])
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df
```

```python
final_df.value_counts("target")
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df["f1"] , final_df["f2"] , c=final_df["target"])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# TRANSFORM THE DATASET
```

```python
oversample = SMOTE()
```

```python
X,y = oversample.fit_resample(final_df[["f1","f2"]] , final_df["target"])
```
