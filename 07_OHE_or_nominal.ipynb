# One Hot Encoder - to represent categorical value into numerical value

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame({"color" : ["red","blue","green","green","red","blue"]})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded_array = encoder.fit_transform(df[["color"]]).toarray()
```

```python
encoded_array
```

```python
encoder_df = pd.DataFrame(encoded_array , columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
encoder.transform([["blue"]]).toarray()   # for new data
```
