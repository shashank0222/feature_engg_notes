```python
import numpy as np
from scipy import stats
```

```python
arr = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
```

```python
arr
```

```python
minimum,Q1,median,Q3,maximum = np.quantile(arr,[0,0.25,0.5,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = Q3-Q1    # interquantile range
```

```python
IQR
```

```python
lower_fence = Q1- 1.5 * IQR

upper_fence = Q3 + 1.5*IQR
```

```python
lower_fence , upper_fence
```
